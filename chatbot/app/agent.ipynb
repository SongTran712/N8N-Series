```python
from pymilvus import MilvusClient, DataType, FunctionType, Function

schema1 = MilvusClient.create_schema(auto_id = False, enable_dynamic_field=False,)

schema1.add_field(
    field_name="id",  # required even with auto_id
    datatype=DataType.INT64,
    is_primary=True,
)

schema1.add_field(
    field_name="name",  # required even with auto_id
    datatype=DataType.VARCHAR,
    max_length=300
)

schema1.add_field(
    field_name="category",  # required even with auto_id
    datatype=DataType.VARCHAR,
    max_length=300
)

schema1.add_field(field_name="image_retrieve_vector", datatype=DataType.FLOAT_VECTOR, dim = 576)
```

```python
index_params1 = client.prepare_index_params()

index_params1.add_index(field_name="image_retrieve_vector", index_type="IVF_FLAT", # Type of the index to create
    index_name="image_vector_index", # Name of the index to create
    metric_type="COSINE", # Metric type used to measure similarity
    params={
        "nlist": 32, # Number of clusters for the index 
        "nprobe": 4
    },
    mmap_enabled=True,
    )

index_params1.add_index(field_name="name", index_type="AUTOINDEX")
index_params1.add_index(field_name="category", index_type="AUTOINDEX")
```

```python
client.drop_collection(collection_name="product_image")
```

```python
client.create_collection(collection_name="product_image", schema=schema1, index_params=index_params1)
```

```python
import timm
from PIL import Image
import torch
import os
from pyvi import ViTokenizer
from sentence_transformers import SentenceTransformer
from transformers import ViTImageProcessor, ViTForImageClassification

# Force CPU
device = torch.device("cpu")

# Load model and send to CPU
model = timm.create_model(
    'tiny_vit_21m_224.dist_in22k_ft_in1k',
    pretrained=True,
    num_classes=0,  # remove classifier nn.Linear
)
model = model.to(device).eval()

# Get model transforms
data_config = timm.data.resolve_model_data_config(model)
transforms = timm.data.create_transform(**data_config, is_training=False)

inputs = []
t = 0

for i in range(len(df)):
    folder_name = df['Tên'][i]
    folder_path = os.path.join("image", folder_name)

    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith((".jpg", ".jpeg", ".png"))]

    for j, filename in enumerate(image_files):
        image_path = os.path.join(folder_path, filename)
        print(f"Processing: {image_path}")

        try:
            image = Image.open(image_path).convert("RGB")
        except Exception as e:
            print(f"Error loading image {image_path}: {e}")
            continue

        try:
            # Move tensor to CPU
            input_tensor = transforms(image).unsqueeze(0).to(device)

            # Forward pass
            with torch.no_grad():
                features = model.forward_features(input_tensor)
                hinh = model.forward_head(features, pre_logits=True)[0].cpu()  # keep it on CPU

            input_doc = {
                "id": t + 1,
                "image_retrieve_vector": hinh.tolist(),  # convert tensor to list for Milvus
                "name": df["Tên"][i],
                "category": df["Nhóm sản phẩm"][i]
            }
            t += 1

            client.insert(collection_name="product_image", data=input_doc)
        except Exception as e:
            print(f"Error processing image {image_path}: {e}")
```

```python
from urllib.request import urlopen
from PIL import Image
import timm
from pymilvus import connections, MilvusClient, DataType

try:
    # Connect to Milvus
    connections.connect(
        alias="default", 
        host='localhost', 
        port='19530'
    )
    
    client = MilvusClient(uri="http://localhost:19530")

    # Load image
    image = Image.open('image/2.png').convert('RGB')

    # Load model
    model = timm.create_model(
        'tiny_vit_21m_224.dist_in22k_ft_in1k',
        pretrained=True,
        num_classes=0,  # remove classifier nn.Linear
    )
    model = model.eval()

    # Preprocess image
    data_config = timm.data.resolve_model_data_config(model)
    transforms = timm.data.create_transform(**data_config, is_training=False)
    output = model.forward_features(transforms(image).unsqueeze(0))
    hinh = model.forward_head(output, pre_logits=True)[0]
    query_vector = hinh.tolist()

    # Milvus search
    res = client.search(
        collection_name="product_image",
        anns_field="image_retrieve_vector",
        data=[query_vector],
        limit=1,
        output_fields=["name", "category"],
        search_params={
            "params": {
                "radius": 0.6,
                # "range_filter": 0.6
            }
        }
    )

    for hits in res:
        for hit in hits:
            print(hit['entity']['category'] )

except FileNotFoundError as e:
    print(f"File not found: {e}")
except ConnectionError as e:
    print(f"Connection failed: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```
